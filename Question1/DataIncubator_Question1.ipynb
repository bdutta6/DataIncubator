# This code analyses data provided by EPA on disinfectants

```python
#Load all the modules

import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
mp.rc('lines', lw=2, c='r')
```

```python
#Load the data
disinfectants_df=pd.read_csv('Data/covid disinfectants.csv')
```

```python
# Check the data
disinfectants_df.head()
```

## Find the average contact time for each formulation type and plot it out

```python
Average_contact_time_df=disinfectants_df.groupby('Formulation Type')['Contact Time (in minutes)'].mean()
#Average_contact_time_list_df=disinfectants_df.groupby('Formulation Type')
```

```python
Average_contact_time_df
```

```python
ax1 = disinfectants_df.plot.scatter(x='Formulation Type',
                      y='Contact Time (in minutes)',
                      c='DarkBlue',
                      figsize=(30,15),
                      linewidth=2,
                      fontsize=24)
ax1.yaxis.label.set_size(40)
ax1.xaxis.label.set_size(40)
```

```python
ax2 = Average_contact_time_df.plot(x='Formulation Type',
                     y='Contact Time (in minutes)',
                     c='Red',
                     figsize=(35,15),
                     linewidth=2,
                     fontsize=24)
ax2.yaxis.label.set_size(40)
ax2.xaxis.label.set_size(40)
```

## Asset 1: The above analysis shows that Towelettes on an average are the fastest acting disinfectants

```python
import seaborn as sns
plt.figure(figsize=(30,10))
plt.xticks(rotation=45)
plt.xticks(fontsize=14)
ax4 = sns.countplot(x="Active Ingredient(s)",data=disinfectants_df)
ax4.yaxis.label.set_size(30)
ax4.xaxis.label.set_size(30)
#disinfectants_df['Formulation Type'].value_counts().plot
```
